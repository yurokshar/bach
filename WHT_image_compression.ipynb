```python
import pathlib

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
from src.compression import compress_rgb_image, restore_rgb_image
```

```python
compress_rgb_image('./resources/lena_color.tiff')
restored = restore_rgb_image('./resources/lena_color.tiff.pak')
```

```python
compressed_size = pathlib.Path('./resources/lena_color.tiff.pak').stat().st_size
original_size = pathlib.Path('./resources/lena_color.tiff').stat().st_size
print('Compressed image size:', compressed_size)
print('Original image size:', original_size)
print('Compression rate:', compressed_size / original_size)
```

```python
image = Image.open('./resources/lena_color.tiff')
fig, ax = plt.subplots(ncols=2, figsize=(12, 12))
ax[0].imshow(np.array(image))
ax[1].imshow(restored)
```
